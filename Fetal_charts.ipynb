```python
# import dependencies
import pandas as pd
```

```python
# read in csv file
data = pd.read_csv("Data/Fetal_Deaths_2005-2019.csv")
data.head()
```

```python
# create comprehensive data frame
data_df = pd.DataFrame(data)
data_df.head()
```

```python
# dropping unneeded columns
data_df = data_df.drop(columns =['Notes', 'Standard Residence States Code', 'Delivery Place Code', 'Delivery Method Code'])
data_df
```

```python
# dropping null rows where all fields are null
data_df = data_df.dropna(how = 'all')
data_df
```

```python
# rename columns
data_df = data_df.rename(columns={
    "Standard Residence States": "state",
    "Mother's Bridged Race/Hispanic Origin": "race_hispanic_origin",
    "Mother's Bridged Race/Hispanic Origin Code": "race_code",
    "Age of Mother Year": "age_mother",
    "Age of Mother Year Code": "age",
    "Delivery Place": "delivery_place",
    "Delivery Method": "delivery_method",
    "Fetal Deaths": "fetal_deaths",
    "Percent of Total Deaths": "percent_deaths"
})
data_df
```

```python
# drop columns now that they have been renamed
data_df = data_df.drop(columns =['race_code','percent_deaths'])
data_df                           
```
